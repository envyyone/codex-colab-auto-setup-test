## 2) Import helper functions


```python
from src.simple_analysis import (
    fit_linear_model,
    generate_fake_data,
    predict_rate,
    save_plot,
)
```

## 3) Generate fake data


```python
df = generate_fake_data(seed=42)
df
```

## 4) Fit a simple linear model


```python
model = fit_linear_model(df)
a = model.coef_[0]
b = model.intercept_
print(f"rate = {a:.4f} * temperature_C + {b:.4f}")
```

## 5) Predict reaction rate at 500 °C


```python
pred_500 = predict_rate(model, 500)
print(f"Predicted rate at 500 °C: {pred_500:.3f}")
```

## 6) Save the plot


```python
output_path = "outputs/temperature_vs_rate.png"
save_plot(df, model, output_path)
print(f"Saved plot to: {output_path}")
```
