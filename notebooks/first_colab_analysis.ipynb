## 1) Environment setup (supports private repositories)

- Create a **GitHub fine-grained personal access token** with **read-only access** to this repository.
- In Colab, open the **Secrets** panel and add a secret named **`GITHUB_TOKEN`**.
- Enable notebook access for that secret.


```python
import os
import sys
from pathlib import Path

REPO_OWNER = "envyyone"
REPO_NAME = "codex-colab-auto-setup-test"
REPO_DIR = Path(f"/content/{REPO_NAME}")

if "google.colab" in sys.modules:
    from google.colab import userdata

    token = userdata.get("GITHUB_TOKEN")
    if not token:
        raise ValueError(
            "Missing GITHUB_TOKEN. In Colab, open the Secrets panel, add a secret named GITHUB_TOKEN, "
            "paste a GitHub fine-grained personal access token with read access to this repo, "
            "and enable notebook access."
        )

    if not REPO_DIR.exists():
        repo_url = f"https://{token}@github.com/{REPO_OWNER}/{REPO_NAME}.git"
        !git clone {repo_url} {REPO_DIR}
    %cd {REPO_DIR}
else:
    repo_root = Path.cwd()
    while repo_root != repo_root.parent and not (repo_root / "requirements.txt").exists():
        repo_root = repo_root.parent
    os.chdir(repo_root)

repo_root = Path.cwd()
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

!pip install -q -r requirements.txt
print(f"Working directory: {repo_root}")

```

## 2) Import helper functions


```python
from src.simple_analysis import generate_fake_data, fit_linear_model, predict_rate, save_plot
```

## 3) Generate fake data


```python
df = generate_fake_data(seed=42)
df
```

## 4) Fit a simple linear model


```python
model = fit_linear_model(df)
a = model.coef_[0]
b = model.intercept_
print(f"rate = {a:.4f} * temperature_C + {b:.4f}")
```

## 5) Predict reaction rate at 500 °C


```python
pred_500 = predict_rate(model, 500)
print(f"Predicted rate at 500 °C: {pred_500:.3f}")
```

## 6) Save the plot


```python
output_path = "outputs/temperature_vs_rate.png"
save_plot(df, model, output_path)
print(f"Saved plot to: {output_path}")
```
